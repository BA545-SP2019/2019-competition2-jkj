```python
import pandas as pd
import numpy as np
```

```python
# Importing Excel file, calling the DataFrame comp_df
df = pd.read_csv('Final5', header=0, index_col=0)
# Checking import 
df.head(5)
```

```python

df = df.astype(int)
df.info()
```

```python
df['Y'].value_counts()
```

```python
df_shuffle=df.iloc[np.random.permutation(len(df))]
df_reset=df_shuffle.reset_index(drop=True)
df_reset.head()
```

```python
df_class = df_reset['Y'].values
```

```python
pd.isnull(df_reset).any()
```

```python
from sklearn.model_selection import train_test_split
training_indices, testing_indices = train_test_split(df_reset.index,
                                                        stratify = df_class,
                                                        train_size=0.75, test_size=0.25, random_state = 2019)
```

```python
training_indices.size, testing_indices.size
```

```python
from tpot import TPOTClassifier
from tpot import TPOTRegressor

#tpot = TPOTClassifier(generations=5,verbosity=2)

#tpot.fit(df_reset.drop('Y',axis=1).loc[training_indices].values,
      #   df_reset.loc[training_indices,'Y'].values)
```

```python
tpot = TPOTClassifier(verbosity=2, max_time_mins=2, max_eval_time_mins=0.04, population_size=20, generations=5, scoring='roc_auc')
tpot.fit(df_reset.drop('Y',axis=1).loc[training_indices].values, df_reset.loc[training_indices,'Y'].values)
```
